```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, C):
    return C*x*np.exp(-x)
```

```python
x_axis= np.linspace(0,10,1000)
C = 1

fig1, ax1 = plt.subplots()
ax1.plot(x_axis, f(x_axis,C))
ax1.set_title(r'$C \cdot x \cdot \exp{(x)}$ with $C = $'+f'{C}')
plt.show()
```

```python
# Define a reasonable range to plot in:
N_points = 10000

xmin,xmax = 0,10
ymin,ymax = 0,0.4
if ymin > ymax:
    ymin,ymax = ymax,ymin


# Accept-reject
np.random.seed(42)
x_uniform = np.random.uniform(size=N_points)*(xmax-xmin) + xmin
y_uniform = np.random.uniform(size=N_points)*(ymax-ymin) + ymin

x_exp = x_uniform[f(x_uniform, C) > y_uniform]
print(len(x_exp))
# Only use the first 1000 generated data points, because that's what the assignment asks for
x_exp = x_exp[:1000]
print(len(x_exp))
```

```python
N_bins = 20
my_range = (-0.25, N_bins/2+0.25)

fig2, ax2 = plt.subplots()

counts,bin_edges,garbage = ax2.hist(x_exp,bins=N_bins+1,range=my_range,label='Generated data')

ax2.plot(x_axis, 500*f(x_axis,C), label=r'$500 \cdot f(x)$',color='black')

median = np.median(x_exp)
ax2.plot([median,median],[0,np.max(counts)],label=f'Median = {median:.3f}',color='red')

ax2.legend()
ax2.set_title(r'$f(x) = C \cdot x \cdot \exp{(x)}$ with $C = $'+f'{C}')
plt.savefig('Problem 3b.png',dpi=600)
#plt.show()
```
